```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (int): The type of conversion selected by user.
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:   # meters to feet
        return value * 3.28084
    elif conversion_type == 2: # feet to meters
        return value / 3.28084
    elif conversion_type == 3: # kg to pounds
        return value * 2.20462
    elif conversion_type == 4: # pounds to kg
        return value / 2.20462
    elif conversion_type == 5: # liters to gallons
        return value * 0.264172
    elif conversion_type == 6: # gallons to liters
        return value / 0.264172
    else:
        return None


#  Main Program
print("Unit Converter")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = int(input("Choose conversion type (1-6): "))
    value = float(input("Enter value to convert: "))

    result = convert_units(choice, value)

    if result is None:
        print("Invalid conversion choice.")
    else:
        print("Converted value:", result)

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

```python
#  Functions
def compute_sum(numbers):
    """
    Returns the sum of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: Sum of the numbers.
    """
    return sum(numbers)


def compute_average(numbers):
    """
    Returns the average of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: Average of the numbers.
    """
    return sum(numbers) / len(numbers)


def compute_max(numbers):
    """
    Returns the maximum number in the list.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: Maximum value in the list.
    """
    return max(numbers)


def compute_min(numbers):
    """
    Returns the minimum number in the list.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: Minimum value in the list.
    """
    return min(numbers)


# Main Program
print("Math Operations on a List")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = int(input("Choose an operation (1-4): "))

    # Take input as a string and convert to a list of floats
    numbers_str = input("Enter numbers separated by spaces: ")
    numbers = [float(x) for x in numbers_str.split()]

    if len(numbers) == 0:
        print("Error: The list is empty.")
    else:
        if choice == 1:
            print("Sum:", compute_sum(numbers))
        elif choice == 2:
            print("Average:", compute_average(numbers))
        elif choice == 3:
            print("Maximum:", compute_max(numbers))
        elif choice == 4:
            print("Minimum:", compute_min(numbers))
        else:
            print("Invalid operation choice.")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

```python
def flatten(lst):
  """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): A nested list (list of lists)

    Returns:
    list: Flattened list containing all elements in a single dimension
  """
  return [i for j in lst for i in j]

nested_list = [[1, 2], [3, 4], [5]]
flattened_list = flatten(nested_list)

print('Original nested list:',nested_list)
print('Flattened list:',flattened_list)



```

Write a Python function that extracts a specific element from a nested list given its indices.
• Requirements:
– Define a function access nested element(lst, indices) that takes a nested list lst and
a list of indices indices, and returns the element at that position.
– Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1,
2], the output should be 6.

```python
def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list using a list of indices.

    Parameters:
    lst (list): The nested list
    indices (list): A list of indices specifying the path to the element

    Returns:
    element: The element at the specified position
    """
    element = lst
    for idx in indices:
        element = element[idx]  # go one level deeper
    return element


# Example
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]  # access lst[1][2]
result = access_nested_element(nested_list, indices)

print("Nested list:", nested_list)
print(f"Element at indices {indices}:", result)
```

```python
def access_nested_element(lst, indices):
    """
    Access an element in a nested list using a list of indices.
    """
    for i in indices:
        lst = lst[i]
    return lst


nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)

print("Nested list:", nested_list)
print("Element at indices", indices, ":", result)
```

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, regardless of depth.

    Parameters:
    lst (list): Nested list containing numbers or other lists

    Returns:
    int/float: Sum of all numbers in the nested list
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total



nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): Nested list
    elem: Element to remove

    Returns:
    list: Nested list with elem removed
    """
    result = []
    for i in lst:
        if isinstance(i, list):
            # Recursively process the sublist
            result.append(remove_element(i, elem))
        else:
            # Only keep i that are not equal to elem
            if i != elem:
                result.append(i)
    return result



nested_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2

new_list = remove_element(nested_list, element_to_remove)
print("Original list:", nested_list)
print("Modified list:", new_list)
```

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    int/float: Maximum element
    """
    max_value = float('-inf')  # smallest possible starting value

    for item in lst:
        if isinstance(item, list):
            # Recursively find max in sublist
            sub_max = find_max(item)
            max_value = max(max_value, sub_max)
        else:
            max_value = max(max_value, item)

    return max_value


# Example
nested_list = [[1, 2], [3, [4, 5]], 6]
print(find_max(nested_list))  # Output: 6
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): Nested list
    elem: Element to count

    Returns:
    int: Number of occurrences of elem
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            # Recursively count in sublist
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1

    return count


# Example
nested_list = [[1, 2], [2, 3], [2, 4]]
element_to_count = 2

print(count_occurrences(nested_list, element_to_count))  # Output: 3
```

```python
def deep_flatten(lst):
    """
    Flattens a nested list of any depth into a single list.

    Parameters:
    lst (list): Nested list

    Returns:
    list: Flattened list
    """
    flat_list = []

    for item in lst:
        if isinstance(item, list):

            flat_list.extend(deep_flatten(item))
        else:

            flat_list.append(item)

    return flat_list


# Example
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(nested_list))
```

```python
def average_nested(lst):
    # Flatten the list recursively
    flat_list = []

    for item in lst:
        if isinstance(item, list):
            flat_list.extend(average_nested(item))  # recursive flatten
        else:
            flat_list.append(item)

    return flat_list  # here we return flattened list first
```

```python
nested_list = [[1, 2], [3, 4], [5, 6]]

flat = average_nested(nested_list)
avg = sum(flat) / len(flat)

print(avg)
```
