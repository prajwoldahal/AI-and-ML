# 3. Image Compression and Decompression using PCA.

```python
# 1. Loading and prepare data
img = Image.open('/content/cat.png').convert('L')
img_arr = np.array(img)

# centering the data
mean = np.mean(img_arr, axis=0)
img_arr_centered = img_arr - mean

# calculating the covariance matrix
cov_matrix = np.cov(img_arr_centered, rowvar=False)
```

```python
# 2. Eigen decomposition and Identifying Principal Components
# calculating eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# sorting eigenvalues in descending order and get the indices
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

k = 50
principal_components = sorted_eigenvectors[:, :k]

# principle components
print(principal_components)

# print
print(principal_components.shape)
```

```python
# Calculate the explained variance ratio
explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

# Calculate the cumulative sum of explained variance ratio
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plot the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative sum plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# so our choice of 50 PC is just about right as per our need.
```

```python
# 3. Reconstruction and Experiment
# Reconstructing with eigen vectors(pcs) to obtain a lower dim rep.
projected_data = np.dot(img_arr_centered, principal_components)

# Reconstruct the image from the projected data
reconstructed_data_full = np.dot(projected_data, principal_components.T) + mean

# Ensure pixel values are within 0-255 range and convert to uint8
reconstructed_data_full = np.clip(reconstructed_data_full, 0, 255).astype(np.uint8)
```

```python
plt.imshow(reconstructed_data_full, cmap='gray')
plt.title(f'Reconstructed Image with k={k} Principal Components')
plt.show()
```
